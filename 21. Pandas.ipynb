```python
result = {"Name":["Areeba", "Abdul Rehman", "Eman", "Abdullah", "Abdul Subhan", "Ummi"],
          "Roll No":[2001, 2002, 2004, 2005, 2007, 2017],
          "Math":[88, 92, 76, 62, 76, 74],
          "Chemistry":[70, 67, 69, 56, 78, 76],
          "Physics":[76, 72, 71, 66, 95, 93],
          "Computer":[76, 74, 69, 61, 74, 80],
          "English":[71, 65, 62, 59, 80, 78]}

df_3 = pd.DataFrame(result)
df_3
```

```python
df_3["Roll No"]
```

```python
df_3[df_3["Math"] >= 80][["Name", "Roll No", "Math"]]
```

```python
df_3[df_3["Roll No"] == 2007]
```

```python
df_3[df_3["Name"].str.startswith(("e", "E"))][["Physics", "English", "Name"]]
```

```python
df_3["Total"] = df_3["Math"] + df_3["Chemistry"] + df_3["Physics"] + df_3["Computer"] + df_3["English"]
df_3
```

```python
df_3["Percentage"] = df_3["Total"] * 100 / 500
df_3
```

```python
def Grade(per):
    grade = None

    if per >= 80:
        grade = "A+"

    elif per >= 70:
        grade = "A"

    elif per >= 60:
        grade = "B"

    elif per >= 50:
        grade = "C"

    elif per >= 40:
        grade = "D"

    else:
        grade = "Fail"

    return grade

df_3["Grade"] = [Grade(per) for per in df_3["Percentage"]]
df_3
```

```python
df_3[df_3["Percentage"] > 75][["Name", "Roll No", "Percentage", "Grade"]]
```

```python
df_3["Percentage"].max()
```

```python
df_3[df_3["Percentage"] == df_3["Percentage"].max()][["Name", "Roll No", "Percentage", "Grade"]]
```

```python
df_3["Percentage"].min()
```

```python
df_3[df_3["Percentage"] == df_3["Percentage"].min()][["Name", "Roll No", "Percentage", "Grade"]]
```
